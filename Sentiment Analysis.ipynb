This is a project concerning the analysis of sentiments regarding Bitcoin and it's relationship to the price.

In a first step we pulled data from Factiva with a lot of articles containing the keyword "Bitcoin" from the past year.
The data could only be pulled as HTML files - 100 articles at a time.

In a first step. We will have to transform the data into a JSON format in order to better be able to handle it for the remainder of the project.

```python
# import statements
import os # file-system interaction
import re # regular expressions
```

```python
# define the regular expressions to find the cutoff points needed later.
reArticlesStart = re.compile(r'<div id="article-')
reArticlesEnd = re.compile(r'<span><div id="pageFooter">')
```

```python
# creates a list of the files in the 
strShardsPath = os.path.join(os.getcwd(), 'shards')
lsShards = os.listdir(strShardsPath)
```

```python
# create new, sorted, list of items based only on the length of the prior list!
# Files HAVE to be numbered!!
lsShardsSorted = []
for i in range(1,len(lsShards)):
    lsShardsSorted.append(str(i) + '.html')
    
print(lsShardsSorted)
```

```python
# This part of the program....

# Creates an empty string
totalString = r''

# iterates through the list of shards
for index, shard in enumerate(lsShardsSorted):
    print(index, shard)
    # checks whether the shard is the first or the last
    if index == 0:
        print('First')
        strShardPath = os.path.join(strShardsPath, shard)
        # opens the file
        with open(strShardPath) as file:
            currentFileString = file.read()
            endMatch = reArticlesEnd.search(currentFileString)
            # appends the first part without the closing html and body tags to the total string
            totalString = totalString + currentFileString[:endMatch.start(0)] + 'END-OF-PAGE-1'
            
    elif index == (len(lsShardsSorted)-1):
        print('Last')
        strShardPath = os.path.join(strShardsPath, shard)
        with open(strShardPath) as file:
            currentFileString = file.read()
            StartMatch = reArticlesStart.search(currentFileString)
            # appends the middle parts without opening as well as closing body and html tags - articles only.
            totalString = totalString + currentFileString[StartMatch.start(0):] + 'END-OF-FILE'
    else: 
        print('In between')
        strShardPath = os.path.join(strShardsPath, shard)
        with open(strShardPath) as file:
            currentFileString = file.read()
            endMatch = reArticlesEnd.search(currentFileString)
            StartMatch = reArticlesStart.search(currentFileString)
            # appends the last part including the closing tags but excluding the opening tags
            totalString = totalString + currentFileString[StartMatch.start(0):endMatch.start(0)] + 'END-OF-SANDWICH-FILE' + '<script src="extraction.js"></script>'
        
# creates a new file or rewrites the existing file with the cobbled together html file.
totalFile = open('total.html', 'w')
totalFile.write(totalString)

```
